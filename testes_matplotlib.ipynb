```python
import pandas as pd
import numpy as np
import scipy as sci
import matplotlib.pyplot as plt
```

```python
R1 = [[0,1,2,3,4],
      [0,1.1,1.9,3.3,4]]

R2 = [[0,1,2,3,4],
      [0,0.5,1.1,1.7,2.3]]
```

```python
plt.style.available
```

```python
#Cor pode ser em hex
# melhor usar label= nos argumentos pra caso queira mudar a ordem

# plt.style.use("dark_background")

plt.plot(R1[0],R1[1],
         color="g", linestyle="--", marker=".", markersize=10, label="Resistor 1")

plt.plot(R2[0],R2[1],
         color="b", linestyle="-.", marker="o", linewidth=3, label="Resistor 2")

plt.title("Tensao vs Corrente")
plt.xlabel("Corrente (A)")
plt.ylabel("Tenso (V)")

plt.grid(True)

plt.legend()

plt.tight_layout() #for padding issues

plt.show()
```
